```python
for name, arr in [('a', a), ('b', b), ('c', c)]:
    print(f"Array {name}: Shape={arr.shape}, Ndim={arr.ndim}, Size={arr.size}")
```

```python
print(f"b dtype: {b.dtype}, name: {b.dtype.name}")
b_int = b.astype(int)
```

```python
add_res = a + b
sub_res = a - b
print("\nAddition (a + b):")
print(add_res)

print("\nSubtraction (a - b):")
print(sub_res)
```

```python
mul_res = b * b
div_res = b / b
print("\nElement-wise Multiplication (b * b):")
print(mul_res)

print("\nElement-wise Division (b / b):")
print(div_res)
```

```python
sqrt_res = np.sqrt(b)
print("Array b:")
print(b)

print("\nSquare Root of Array b (sqrt_res):")
print(sqrt_res)
```
